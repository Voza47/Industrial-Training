```python
#histogram
import matplotlib.pyplot as plt

ages=[2,5,70,40,30,65,18,43]
range=(0,100)
bins=10
# plt.hist(ages,bins,histtype="bar",color="Brown",edgecolor="Black")
plt.xlabel("AGE GROUP")
plt.hist(ages,histtype='bar', bins=10,color='blue',alpha=0.5)
plt.ylabel("NUMBER OF PEOPLE")
plt.title('My Histogram')
plt.show()
```

```python
#scatter graph

import matplotlib.pyplot as plt
x_values=[1,2,3,4,5,6,7,8]
y_values=[1,4,9,16,25,36,49,64]

plt.scatter(y_values,x_values,label="stars",s=100,color="blue",edgecolor="Black",marker=".")
plt.show()
```

```python
#pie Chart
import matplotlib.pyplot as plt
#defining labels
activities =['eat','sleep','work','play']
#color for each label
colors=['r','y','g','b']
#label covered for by each label
slices=[3,7,8,6]
#plotting th pie chart
plt.pie(slices,colors=colors,labels=activities,autopct='%1.1f%%',shadow=True,explode=(0,0.1,0,0),radius=1.2,startangle=90)
plt.legend()
```

```python
#Plotting curves of given equation

import matplotlib.pyplot as plt
import numpy as np

# setting the x - coordinates
x=np.arange(0,2*(np.pi),0.1)
#setting the corresponding y-coordinates
y=np.sin(x)

plt.plot(x,y)

plt.show()
```
